#  Logistic Regression Using Gradient Descent 

### Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y =  make_blobs(n_samples=100, centers=[[2,4],[4,2]], random_state=20)
```

```python
#Visualize dataset 
plt.plot(X[:,0][y==0],X[:,1][y==0],'o',color='red')
plt.plot(X[:,0][y==1],X[:,1][y==1],'^',color='blue')
```

```python
w0,w1,w2 = (np.random.rand(3)-0.5)*2
x_axis = np.linspace(0,6)
```

### Initial classifier(random)

```python
#Visualize dataset 
plt.plot(X[:,0][y==0],X[:,1][y==0],'o',color='red')
plt.plot(X[:,0][y==1],X[:,1][y==1],'^',color='blue')
plt.plot(x_axis,-(x_axis*w0+w2)/w1)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def crossentropy(p,x):
    return (-1/X.shape[0])*np.sum((p*np.log(x)) + (1-p)*np.log(1-x))
```

```python
rate = 0.01
cost = np.zeros(100)
for i in range(100):
    cost[i] = crossentropy(y,sigmoid((w0*X[:,0] + w1*X[:,1] + w2) - y))
    for j in range(X.shape[0]):
        output = sigmoid((w0*X[j,0] + w1*X[j,1] + w2) - y[j])
        dw0 =  X[j,0]*(output-y[j])*rate
        dw1 =  X[j,1]*(output-y[j])*rate
        dw2 =  (output-y[j])*rate
        w0 -= dw0
        w1 -= dw1
        w2 -= dw2
```

### Cost:

```python
plt.plot(cost)
plt.show()
```

### Final classifier

```python
#Visualize dataset 
plt.plot(X[:,0][y==0],X[:,1][y==0],'o',color='red')
plt.plot(X[:,0][y==1],X[:,1][y==1],'^',color='blue')
plt.plot(x_axis,-(x_axis*w0+w2)/w1)
plt.show()
```
